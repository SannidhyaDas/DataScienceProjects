```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#we need first 3 cast of all the movies 
movies['cast'][0]     #casts of the first movie Avatar
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
# we need that dictionary where job = director 
movies['crew'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'][0]
#we are going to concatinate this string type column also a list so that when we further proceed we will concat this also with our further list type columns.
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
#Transformations for removing the space b/w two words 
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
#Linear Algebra , creating vectors and placing them in a higher dimensional plane 
# after placing we will choose the closest vectors 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
vector[3]
```

```python
list(cv.get_feature_names_out())
```

```python
#now we will calculate the distance between two vectors, and for this we will don't consdider euclidiean distance
#instead of this we will measure the coshine distance of the two vectors. 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
sorted(similarity[4],reverse = True)      #not so recommended due to we can't match it with movie id , after sorting 
```

```python
sorted(list(enumerate(similarity[3])),reverse = True, key = lambda x:x[1])[1:8]
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:8]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Batman v Superman: Dawn of Justice')
```

```python
recommend('The Avengers')
```

```python
recommend('Gandhi')
```
